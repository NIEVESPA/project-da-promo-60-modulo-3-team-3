#### > **¿De qué manera la frecuencia de viajes laborales, impacta en la economía personal del empleado y su nivel de satisfacción general con la vida personal y laboral?**

```python
df["businesstravel"].unique()
```

```python
df_viajes = df.groupby("businesstravel")["employeenumber"].count().reset_index()
df_viajes #OBSERVAMOS Nº DE EMPLEADOS POR CATEGORÍA
```

columnas a analizar:

1. Frecuencia Viajes
2. Economía Personal: compensación por los trastornos que ocasionan los viajes laborales
3. Satisfacción Laboral: ¿el viaje mejora o deteriora la expaeriencia laboral?
4. Satisfacción Personal: ¿los viajes tienen un costo en la vida personal del empleado?

```python
df[["businesstravel", "monthlyincome($)", "jobsatisfaction", "worklifebalance"]].head(4) #VALORACIÓN POR RANGO[1:4]
```

```python
df[["monthlyincome($)", "jobsatisfaction", "worklifebalance"]].describe().T
```

```python
viajes_vs_economia = df.groupby("businesstravel")["monthlyincome($)"].median().sort_values(ascending=False).reset_index()
print(f'mediana de Ingreso Mensual por frecuencia de viajes:')

viajes_vs_economia
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize = (20,5)) #FILAS Y COLUMNAS QUE QUEREMOS

colores = ["seagreen", "teal", "mediumseagreen", "cadetblue"] #COLORES POR CATEGORÍA
 
axes[0].pie("employeenumber", labels= "businesstravel", #axes[]:INDICAMOS SUBPLOTS
        data = df_viajes, 
        autopct=  '%1.1f%%', 
        colors = colores, 
        textprops={'fontsize': 8}, 
        startangle=30)

axes[0].set_title("Empleados vs viajes", fontsize = 9); #TÍTULO 

sns.countplot(x = "businesstravel", 
              data = df, 
              ax = axes[1],
              hue="joblevel", 
              palette = "viridis")

axes[1].set_xticklabels(axes[1].get_xticklabels())
axes[1].set_xlabel("VIAJES", fontsize = 10)
axes[1].set_ylabel("Nº EMPLEADOS",  fontsize = 9); #FILAS Y COLUMNAS QUE QUEREMOS

sns.barplot(data=df,
            x="joblevel", 
            y = "monthlyincome($)", 
            errorbar=None,
            ax = axes[2],
            color = "seagreen")

axes[2].set_xlabel("NIVEL LABORAL", fontsize = 10)
axes[2].set_ylabel("SALARIO",  fontsize = 9)

plt.tight_layout(); #AJUSTAMOS ELEMENTOS A LA GRAFICA;

sns.countplot(x = "joblevel", 
              data = df, 
              ax = axes[3],
              hue="businesstravel", 
              palette = "viridis")

axes[3].set_xticklabels(axes[3].get_xticklabels())
axes[3].set_xlabel("NIVEL LABORAL", fontsize = 10)
axes[3].set_ylabel("Nº EMPLEADOS",  fontsize = 9); #FILAS Y COLUMNAS QUE QUEREMOS
```

```python
tabla_joblevel = pd.crosstab(
    index=df["businesstravel"],
    columns=df["joblevel"],
    normalize="index"
)*100
print(tabla_joblevel)
```

Nuestra primera observación es que más de la mitad de los empleados de la empresa no viaja. 
Tras analizar la frecuencia de viajes en base al nivel jerárquico de los empleados: 
1. Contrario a lo que podríamos pensar, los niveles más altos no viajan más, en este caso la diferencia no es tan marcada. De echo, el grupo que viajan frecuentemmente tiene un a proporción muy baja en nivel Senior.
2. Observamos una gran concentración de empleados en niveles inferiores en todas las categorías de viajes. Independientemente de la freceiencia de viajes, estos son los niveles jerárquicos con más empleados. 
3. la responsabilidad de viajar es asumida por roles técnicos y no directivos.


```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (20,5)) #FILAS Y COLUMNAS QUE QUEREMOS

sns.boxplot(y = "businesstravel", 
            x = "monthlyincome($)", 
            data = df, 
            ax = axes[0], 
            palette = "viridis")

axes[0].set_xlabel("FRECUENCIA VIAJES")
axes[0].set_ylabel("SALARIO",
                    fontsize = 9)

sns.countplot(x = "businesstravel", 
            hue =  "jobsatisfaction", 
            data = df,
            ax = axes[1],
            color = "teal")

axes[1].set_xlabel("SATISFACCIÓN LABORAL", fontsize = 10)
axes[1].set_ylabel("FRECUENCIA VIAJES",  fontsize = 9)

sns.countplot(x = "businesstravel", 
            hue =  "worklifebalance", 
            data = df,
            ax = axes[2],
            palette = "viridis")

axes[2].set_xlabel("SATISFACCIÓN PERSONAL", fontsize = 10)
axes[2].set_ylabel("FRECUENCIA VIAJES",  fontsize = 9)


plt.tight_layout(); #AJUSTAMOS ELEMENTOS A LA GRAFICA;
```

```python
tabla_jobsatisfaction= pd.crosstab(
    index=df["businesstravel"],
    columns=df["jobsatisfaction"],
    normalize="index"
)*100
print(tabla_jobsatisfaction)
```

```python
tabla_worklifebalance = pd.crosstab(
    index=df["businesstravel"],
    columns=df["worklifebalance"],
    normalize="index"
)*100
print(tabla_worklifebalance)
```

Nuestra segunda observación es la relación entre salario y la frecuencia de viajes de los empleados:
1. La frecuencia de viajes no es un factor determinante de ingreso. El promedio de empleados que viajan frecuentemente no perciben un aumento en su salario, en comparación a otros niveles que viajan rara vez o no viajan. Tal y como hemos visto con el nivel laboral, los salarios más altos se concentran en roles que no requieren viajes constantes.

En tercer lugar, hemos analizado la frecuencia de viajes con su satisfacción laboral y personal:
1. Aunque la mayoria de empleados están satisfechos (nivel 3 y 4), el conteo más alto de empleados insatisfechos lo tenemos en el grupo que no viaja. Lo que no hace entender la frecuencia de viajes como un factor que no concluyente para marcar la insatisfacción laboral. 
2. De igual modo, la satisfacción personal refuerza la idea de que los viajes no son un motivo de insatisfacción. Sin embargo, los empleados de la categoría viajes frecuentes no llegan a una satisfacción promedio nivel 4.

Por último, analizamos si existe un riesgo de rotación futura del personal:

A rasgos generales, la frecuencia de viajes no parece ser un factor decisivo en valoración de estrés o insatisfacción personal.
