#### > **驴Cu谩l es la relaci贸n entre la edad media de los empleados y las tasas derotaci贸n de personal (correlaci贸n entre juventud y mayor rotaci贸n vs. edad avanzada y estabilidad)?**

```python
# 1. Definir los l铆mites de los bins (de 18 a帽os hasta el m谩ximo, en pasos de 5)
bins = range(18, df["age"].max() + 5, 5)

# 2. Crear las etiquetas de los rangos (ej. '18-22', '23-27')
labels = [f"{b}-{b+4}" for b in bins[:-1]]

# 3. Aplicar el corte para crear la nueva columna 'age_bin'
df["age_bin"] = pd.cut(df["age"], bins=bins, labels=labels, right=False)
```

```python
# Crear una nueva columna binaria 'attrition_flag'

df['attrition_flag'] = df['attrition'].apply(lambda x: 1 if x == 'yes' else 0)

# El Por Qu茅: Esto asigna 1 solo cuando el valor es 'yes', y 0 para 'no'
```

```python
# 1. Agrupar por el rango de edad ('age_bin')
# 2. Calcular la media ('mean') de la bandera binaria ('attrition_flag')
# La media de una variable 0/1 es la proporci贸n o TASA.
df_rate = df.groupby("age_bin")["attrition_flag"].agg(tasa_rotacion="mean").reset_index()

# Mostrar el resultado (los rangos de edad con sus respectivas tasas de rotaci贸n)
print("--- Tasa de Rotaci贸n por Rango de Edad ---")
print(df_rate)
```

```python
plt.figure(figsize=(12, 6))

# Usamos 'age_bin' en el eje X y 'tasa_rotacion' en el eje Y
sns.barplot(data=df_rate, x="age_bin", y="tasa_rotacion", palette="viridis")

# Mejorando la est茅tica para una presentaci贸n profesional
plt.title("Tasa de Rotaci贸n (Attrition) por Rango de Edad", fontsize=16, weight='bold')
plt.xlabel("Rango de Edad", fontsize=12)

# Convertimos la etiqueta del Eje Y a porcentaje para que sea m谩s intuitivo para el negocio
plt.ylabel("Tasa de Rotaci贸n (%)", fontsize=12)

# Opcional: Si quieres forzar que el eje Y se muestre como porcentaje:
# max_rate = df_rate['tasa_rotacion'].max() * 1.1 # Buscar el valor m谩ximo para el l铆mite del eje Y
# ticks = np.linspace(0, max_rate, 6) # Crear 6 ticks
# plt.yticks(ticks, [f'{i*100:.1f}%' for i in ticks]) 

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

 Las personas que tienen entre 18 y 22 a帽os son los que presentan una mayor rotaci贸n, con mucha diferencia con respecto a los otros grupos de edad. Siendo los de 38-41 los que menos rotaci贸n presentan. Si que observamos relaci贸n entre juventud y rotaci贸n, aunque la franja de 58-62 presenta mayor rotaci贸n que las personas de entre 38 y 57 a帽os.
