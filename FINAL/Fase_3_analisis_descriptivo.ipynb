La meta no es solo cuantificar la rotación, sino comprender las causas subyacentes a través de la exploración de variables demográficas, organizacionales y de bienestar.

**Objetivos Clave de la Investigación:**

1.	Identificar Factores de Riesgo: Determinar qué variables demográficas (edad, género, estado civil, antigüedad), de compensación (incentivos, salario) y de gestión (liderazgo, promociones) están más fuertemente correlacionadas con la satisfacción y la permanencia.

3.	Optimizar Estrategias de Retención: Generar insights accionables que permitan a la organización diseñar e implementar políticas de retención más precisas, beneficios más atractivos y planes de desarrollo profesional que se adapten a las necesidades específicas de su fuerza laboral.

**En esencia, este análisis busca transformar los datos brutos de recursos humanos en una hoja de ruta estratégica para mejorar el compromiso, la productividad y la estabilidad de la plantilla.**

#### > **¿Cómo influye la edad del empleado en la probabilidad de recibir una promoción, y cuál es el tiempo medio desde la última promoción recibida en función de la edad?**

```python
df[["age", "yearssincelastpromotion"]].describe().T
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(
    data=df,
    x='age',
    bins=18,                 
    kde=True,                
    color='blue',         
    edgecolor='black', 
    linewidth=1.2,           
    alpha=0.7                
)

plt.title('Distribución de Frecuencia de la Edad de los Empleados', fontsize=16)
plt.xlabel('Edad del Empleado (age)', fontsize=12)
plt.ylabel('Frecuencia (Conteo)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

Lo primero que observamos es cómo se distribuye la edad de los empleados. El mayor pico lo encontramos entre los 28 y los 42 años

```python
df["age_cat"] = pd.cut(df["age"], bins=3, labels=['Joven', 'Mediano', 'Adulto'])#hacemos 3 categorias distintas de edad para tener una idea  mas aproximada de como se comportan los años desde la ultima promocion segun la edad
```

```python
df_promocion_media = df.groupby('age_cat')['yearssincelastpromotion'].mean().reset_index().round(2)
df_promocion_media
```

```python
df_promocion_media = df.groupby('age_cat')['yearssincelastpromotion'].mean().reset_index() #calculamos la media de años desde la ultima promocion por edad

plt.plot(
    df_promocion_media['age_cat'],
    df_promocion_media['yearssincelastpromotion'])

# Paso 3: Etiquetas y Títulos
plt.title('Media de Años Desde Última Promoción vs. Edad')
plt.xlabel('Edad')
plt.ylabel('Media de Años Sin Promoción')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(9, 6))

sns.barplot(
    data=df_promocion_media,
    x='age_cat',
    y='yearssincelastpromotion',
    palette='magma', # Paleta de colores cálidos y degradados (Rojo/Burdeos)
    edgecolor='black',
    linewidth=0.8
)

# Estética y Etiquetas
plt.title('Media de Años Sin Promoción por Categoría de Edad ', fontsize=16, color='#580F41') # Color Burdeos oscuro
plt.xlabel('Categoría de Edad (age_cat)', fontsize=12)
plt.ylabel('Media de Años Desde Última Promoción', fontsize=12)
plt.ylim(0) # Siempre empezar en 0 para gráficos de barras para evitar distorsión
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Optamos por dividir la edad por rangos para así poder observar si hay una diferencias a la hora de otorgar las promociones:

-El primer rango de edad se encuntra entre los 20 y los 34 años. Este es un rango que se caracteriza por una rápida progresión en el trabajo, siendo más constantes las promociones, con una media de 1.46 años.

-El segundo rango se encuentra entre los 35 y los 42 años. En esta etapa se tiene más experiencia y se produce una cierta consolidación, permitiendo que accedan a roles de coordinación. Aquí la media de años por promoción es de 2.48 años.

-El tercer rango está comprendido entre los 43 y los 62 años. En este punto, las carreras suelen estar más consolidadas y se puede producir el estancamiento de las mismas, al igual que se puede dar una posible discriminación por edad. La media de años por promoción es superior al resto, siendo esta de 3.27 años.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['attrition']=='no'], x='yearssincelastpromotion', fill=True, label='Se queda', color='blue')
sns.kdeplot(data=df[df['attrition']=='yes'], x='yearssincelastpromotion', fill=True, label='Se va', color='magenta')
plt.title('Años desde la Última Promoción')
plt.xlabel('Años')
plt.legend()
plt.show()
```

Ambas curvas alcanzan su punto máximo cerca de 0 años (es decir, en el año actual o el año pasado). Esto es normal debido a que los empleados de la empresa tendrán una promoción reciente o habrán pasado poco tiempo desde la última.

Sin embargo, observamos que a pesar de que la promoción ha sido reciente, no es suficiente para retener a los empleados, bien sea porque no es suficiente la mejora de las condiciones o porque ha utilizado esta promoción para pedir mejores condiciones en otras empresas.

A partir de aquí, vemos que se produce un estancamiento en el crecimiento laboral, lo que puede explicar el ligero crecimiento en la curva de los empleados que deciden irse. De igual manera, la cantidad de empleados que deciden quedarse a pesar de este estancamiento puede estar justificado por el nivel de beneficios que obtienen en la empresa o simplemente por inercia. (Sería interesante ver qué nivel de satisfacción tienen estos empleados para así evitar futuras rotaciones)

```python
bins_propuestos = [20, 35, 43, 63]
labels_propuestos = ['20-34', '35-42', '43-62']
df['age_cat'] = pd.cut(df['age'], bins=bins_propuestos, labels=labels_propuestos, right=False)

df['attrition_num'] = df['attrition'].apply(lambda x: 1 if x == 'Yes' else 0)

fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# --- GRÁFICO 1: DISTRIBUCIÓN DE EDAD (Histograma) ---
sns.histplot(
    data=df,
    x='age',
    bins=18,
    kde=True,
    color='skyblue',
    edgecolor='black',
    linewidth=1.2,
    alpha=0.7,
    ax=axes[0]  
)
axes[0].set_title('1. Distribución de Edad', fontsize=14)
axes[0].set_xlabel('Edad (age)', fontsize=10)
axes[0].set_ylabel('Frecuencia (Conteo)', fontsize=10)
axes[0].grid(axis='y', linestyle='--', alpha=0.5)


# --- GRÁFICO 2: MEDIA DE PROMOCIÓN POR CATEGORÍA DE EDAD (Barplot) ---
sns.barplot(
    data=df_promocion_media,
    x='age_cat',
    y='yearssincelastpromotion',
    palette='magma',
    edgecolor='black',
    linewidth=0.8,
    ax=axes[1] 
)
axes[1].set_title('2. Media Años Sin Promoción por Categoría', fontsize=14, color='#580F41')
axes[1].set_xlabel('Categoría de Edad (age_cat)', fontsize=10)
axes[1].set_ylabel('Media de Años Desde Última Promoción', fontsize=10)
axes[1].set_ylim(0)
axes[1].grid(axis='y', linestyle='--', alpha=0.6)


# --- GRÁFICO 3: ROTACIÓN VS. PROMOCIÓN (KDE Plot) ---
sns.kdeplot(
    data=df[df['attrition']=='no'],
    x='yearssincelastpromotion',
    fill=True,
    label='Se queda',
    color='blue',
    ax=axes[2] 
)
sns.kdeplot(
    data=df[df['attrition']=='yes'], 
    x='yearssincelastpromotion',
    fill=True,
    label='Se va',
    color='magenta',
    ax=axes[2]
)
axes[2].set_title('3. Años desde la Última Promoción vs. Rotación', fontsize=14)
axes[2].set_xlabel('Años desde la Última Promoción', fontsize=10)
axes[2].set_ylabel('Densidad', fontsize=10)
axes[2].legend(loc='upper right')

plt.suptitle('Análisis de Equidad y Progresión de Carrera', fontsize=18, y=1.02)
plt.tight_layout(rect=[0, 0, 1, 0.98]) 
plt.show()
```
