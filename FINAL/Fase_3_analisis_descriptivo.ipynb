```python
df_promocion_media = df.groupby('age_cat')['yearssincelastpromotion'].mean().reset_index().round(2)
df_promocion_media
```

```python
plt.figure(figsize=(9, 6))

sns.barplot(
    data=df_promocion_media,
    x='age_cat',
    y='yearssincelastpromotion',
    palette='magma', # Paleta de colores cálidos y degradados (Rojo/Burdeos)
    edgecolor='black',
    linewidth=0.8
)

# Estética y Etiquetas
plt.title('Media de Años Sin Promoción por Categoría de Edad ', fontsize=16, color='#580F41') # Color Burdeos oscuro
plt.xlabel('Categoría de Edad (age_cat)', fontsize=12)
plt.ylabel('Media de Años Desde Última Promoción', fontsize=12)
plt.ylim(0) # Siempre empezar en 0 para gráficos de barras para evitar distorsión
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Optamos por dividir la edad por rangos para así poder observar si hay una diferencias a la hora de otorgar las promociones:

-El primer rango de edad se encuntra entre los 20 y los 34 años. Este es un rango que se caracteriza por una rápida progresión en el trabajo, siendo más constantes las promociones, con una media de 1.46 años.

-El segundo rango se encuentra entre los 35 y los 42 años. En esta etapa se tiene más experiencia y se produce una cierta consolidación, permitiendo que accedan a roles de coordinación. Aquí la media de años por promoción es de 2.48 años.

-El tercer rango está comprendido entre los 43 y los 62 años. En este punto, las carreras suelen estar más consolidadas y se puede producir el estancamiento de las mismas, al igual que se puede dar una posible discriminación por edad. La media de años por promoción es superior al resto, siendo esta de 3.27 años.

#### **¿Cuál es el nivel de satisfacción general de los empleados y cómo se correlaciona con la media de edad en los distintos departamentos o equipos?**

- **Nivel de satisfacción por categoria de ex-empleados vs empleados**

```python
df_yes = df[df['attrition'] == 'yes']
df_no = df[df['attrition'] == 'no']
```

```python
cols = ['jobsatisfaction', 'environmentsatisfaction', 'relationshipsatisfaction', 'worklifebalance']


# Calcular promedio para cada grupo
avg_yes = round(df_yes[cols].mean(), 2).reset_index()
avg_yes.columns = ['Satisfaction', 'Mean']
avg_yes['attrition'] = 'Ex-empleados'


avg_no = round(df_no[cols].mean(), 2).reset_index()
avg_no.columns = ['Satisfaction', 'Mean']
avg_no['attrition'] = 'Empleados'

# Combinar en un solo DataFrame
avg_sat_df = pd.concat([avg_yes, avg_no], ignore_index=True)

# Graficar
plt.figure(figsize=(10,4))
sns.barplot(y='Satisfaction', x='Mean', hue='attrition', data=avg_sat_df, palette='YlGn_r')
plt.title('SATISFACCIÓN SEGÚN ROTACIÓN')
plt.xlabel('promedio')
plt.ylabel('')
plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left', borderaxespad=0.)
plt.tight_layout()
plt.show()

# Seleccionamos las columnas de interés
cols = ['jobsatisfaction', 'worklifebalance', 'environmentsatisfaction', 'relationshipsatisfaction', 'attrition']

# Creamos un DataFrame “largo”
df_long = df[cols].melt(
    id_vars='attrition',
    var_name='Satisfaction_Type',
    value_name='Score')

# Gráfico
plt.figure(figsize=(10,6))
sns.boxplot(
    x='Satisfaction_Type',
    y='Score',
    hue='attrition',
    data=df_long,
    palette='YlGn')
plt.title('Comparación de distintos tipos de satisfacción según Attrition')
plt.ylabel('Satisfacción')
plt.xlabel('Tipo de satisfacción')
plt.legend(title='Attrition')
plt.show()
```

**CONCLUSIÓN**
- Se puede observar que los empleados que dejaron la empresa estaban menos satisfechos con el trabajo y el ambiente laboral.

- **Relación de satisfacción laboral con la distancia al trabajo, empleados vs ex-empleados**

```python
# --- Crear rangos de distancia ---
bins_b = [0, 5, 10, 15, 20, 25, 30, 35]  # ajusta según tus datos
labels_l = ["0-5","6-10","11-15","16-20","21-25","26-30","31-35"]

df_yes["dist_range"] = pd.cut(df_yes["distancefromhome"], bins=bins_b, labels=labels_l)
df_no["dist_range"]  = pd.cut(df_no["distancefromhome"], bins=bins_b, labels=labels_l)

# --- Calcular promedio de jobsatisfaction por rango ---
jobsat_yes = df_yes.groupby("dist_range")["jobsatisfaction"].mean()
jobsat_no  = df_no.groupby("dist_range")["jobsatisfaction"].mean()

# --- Crear DataFrame para graficar ---
df_jobsat_distance = pd.DataFrame({
    "ex-empleados": jobsat_yes,
    "empleados": jobsat_no
})

# --- Gráfico de barras con línea de tendencia ---
plt.figure(figsize=(10, 6))
x = np.arange(len(df_jobsat_distance)) 

# Barras
plt.bar(x - 0.15, df_jobsat_distance["ex-empleados"], width=0.3, color="#3D1D1D", label="ex-empleados")
plt.bar(x + 0.15, df_jobsat_distance["empleados"], width=0.3, color="#f1aa6d", label="empleados")

plt.xticks(x, df_jobsat_distance.index)
plt.xlabel("distancia en millas")
plt.ylabel("promedio satisfacción")
plt.title("SATISFACCIÓN LABORAL vs DISTANCIA AL TRABAJO")

# --- Línea de tendencia para ex-empleados ---
coef_yes = np.polyfit(x, df_jobsat_distance["ex-empleados"], 1)  # regresión lineal
trend_yes = np.poly1d(coef_yes)
plt.plot(x, trend_yes(x), color="#3D1D1D", linestyle="--")

# --- Línea de tendencia para empleados ---
coef_no = np.polyfit(x, df_jobsat_distance["empleados"], 1)
trend_no = np.poly1d(coef_no)
plt.plot(x, trend_no(x), color="#f1aa6d", linestyle="--")

plt.legend(title="", bbox_to_anchor=(1.01, 1), loc="upper left")
plt.tight_layout()
plt.show()
```

**CONCLUSIÓN**
- **La suposición de que la opción de trabajo remoto e la distancia del trabajo pudieran ser uno de los factores que influye en la salida de la empresa no se verifica. Estos no son factores que estén llevando a la rotación (en este caso).**
