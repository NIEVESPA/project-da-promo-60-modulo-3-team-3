**stock options**
- son incentivos que permiten a los empleados comprar acciones de la empresa a un precio fijo (generalmente más bajo que el de mercado) después de cierto tiempo o cumpliendo ciertos objetivos.

- es un indicador de cuánto puede beneficiarse un empleado de las stock options dentro de la empresa.



    0 → No tiene asignadas stock options.

    1 → Nivel básico, pocas opciones asignadas.

    2 → Nivel intermedio, más opciones que el nivel 1.

    3 → Nivel alto, la mayor cantidad de opciones asignadas, normalmente para puestos clave o ejecutivos.

En otras palabras, a mayor número, mayor beneficio potencial de las stock options.

NOTAS ESTRATEGIAS:

1. OPCIÓN1 <Corrección por Valor Absoluto>:
podemos asumir que el signo negativo se trata de un error de entrada de la información. solución, convertir todos los valores negativos a su valor absoluto positivo. Adecuada si el número de casos es pequeño.
2. OPCIÓN2 <Imputación por mediana o media>:
sería una opción si consideramos que la calidad del dato es cuestionable. La mediana es más robusta y clave pera la distancia.
3. OPCIÓN3 <Eliminación Filas>:
no contemplamos esta opción, no queremos arriesgar el análisis. Puede impedir entender el conjunto de datos o romper la continuidad del índice.

CONCLUSIÓN:

Dado que el porcentaje de negativos es 12%, y teniendo en cuenta que es un porcentaje alto, preservamos la integridad de la información original y elegimos la opción1. Corregimos signo, convirtiendo a Valor absoluto, ya la imputación significaría un 12% de los datos.
